```python
dummies = pd.get_dummies(df3, columns = [  'fueltype', 'aspiration', 'doornumber',
                                        'carbody', 'enginelocation','enginetype', 'cylindernumber',
                                        'fuelsystem', 'peakrpm'
                                                        ],
                              drop_first=True)
dummies.head(5)
```

```python
df4 = df3.drop(['fueltype', 'aspiration', 'doornumber','carbody', 'enginelocation','enginetype', 'cylindernumber','fuelsystem', 'peakrpm'], axis='columns')
df4.head()
```

```python
df5 = pd.concat([df4,dummies],axis='columns')
df5.head(5)
```

```python
df5.shape
```

```python
X = df5.drop(['price','CarName'],axis='columns')
X.head(3)
```

```python
X.shape
```

```python
y = df5.price
y.head(3)
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
#i will fix this porsion later 


def predict_price(mpg,horsepower,symboling,enginesize):    
    loc_index = np.where(X.columns==mpg)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = horsepower
    x[1] = symboling
    x[2] = enginesize 
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```
